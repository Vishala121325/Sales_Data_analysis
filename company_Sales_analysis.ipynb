# Automobile Sales Analysis Assignment
This notebook contains all the required tasks for visualizing and analyzing automobile sales data using Matplotlib, Seaborn, Pandas, and Dash.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## TASK 1.1

```python
# Read data
df = pd.read_csv("company_sales_data.csv")  # Replace with actual file name

# Exercise 1: Line plot of total profit
plt.figure(figsize=(10,5))
plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Company Profit Per Month')
plt.show()
```

## TASK 1.2

```python
# Exercise 2: Styled line plot
plt.figure(figsize=(10,5))
plt.plot(df['month_number'], df['total_units'], color='red', linestyle='dotted', linewidth=3,
         marker='o', markerfacecolor='red', label='Units Sold')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
plt.title('Monthly Units Sold')
plt.show()
```

## TASK 1.3

```python
# Exercise 3: Multi-line plot
plt.figure(figsize=(12,6))
products = df.columns[1:-2]
for product in products:
    plt.plot(df['month_number'], df[product], label=product)
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.title('Monthly Sales for Each Product')
plt.legend()
plt.grid(True)
plt.show()
```

## TASK 1.4

```python
# Exercise 4: Scatter plot for toothpaste
plt.figure(figsize=(10,5))
plt.scatter(df['month_number'], df['toothpaste'], color='blue')
plt.grid(linestyle='--')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste units sold')
plt.title('Toothpaste Sales Data')
plt.show()
```

## TASK 1.5

```python
# Exercise 5: Bar chart for face cream and face wash
plt.figure(figsize=(10,5))
bar_width = 0.4
months = df['month_number']
plt.bar(months - 0.2, df['facecream'], width=bar_width, label='Face Cream')
plt.bar(months + 0.2, df['facewash'], width=bar_width, label='Face Wash')
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.title('Face Cream and Face Wash Sales')
plt.legend()
plt.show()
```

## TASK 1.6

```python
# Exercise 6: Bar chart for bathing soap + save
plt.figure(figsize=(10,5))
plt.bar(df['month_number'], df['bathingsoap'])
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.title('Bathing Soap Sales')
plt.savefig('bathing_soap_sales.png')  # Saves to disk
plt.show()
```

## TASK 1.7

```python
# Exercise 7: Histogram of total profit
plt.figure(figsize=(10,5))
plt.hist(df['total_profit'], bins=5)
plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.title('Profit Distribution Histogram')
plt.show()
```

## TASK 1.8

```python
# Exercise 8: Pie chart of annual sales by product
product_totals = df[products].sum()
plt.figure(figsize=(8,8))
plt.pie(product_totals, labels=products, autopct='%1.1f%%')
plt.title('Total Yearly Sales Distribution')
plt.show()
```

## TASK 1.9

```python
# Exercise 9: Subplot for bathing soap and facewash
fig, axs = plt.subplots(2, 1, figsize=(10,8))
axs[0].plot(df['month_number'], df['bathingsoap'], label='Bathing Soap')
axs[0].set_title('Bathing Soap Sales')
axs[1].plot(df['month_number'], df['facewash'], label='Face Wash', color='orange')
axs[1].set_title('Face Wash Sales')
plt.tight_layout()
plt.show()
```

```python
# Exercise 10: Stack plot
plt.figure(figsize=(10,6))
plt.stackplot(df['month_number'], [df[p] for p in products], labels=products)
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.title('Sales Data - Stack Plot')
plt.legend(loc='upper left')
plt.show()
```
